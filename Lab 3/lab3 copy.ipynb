```python
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes
```

```python
# Desencriptar la imagen con AES-ECB
def decrypt_image_aes_ecb(image_path, key):
    with open(image_path, 'rb') as file:
        data = file.read()
        cipher = AES.new(key, AES.MODE_ECB)
        decrypted_data = cipher.decrypt(data)
        with open(f'decrypted_aes_ecb_{image_path}', 'wb') as file:
            file.write(decrypted_data)
```

```python
def decrypt_image_aes_cbc(image_path, key, iv):
    with open(image_path, 'rb') as file:
        data = file.read()
        cipher = AES.new(key, AES.MODE_CBC, iv)
        decrypted_data = cipher.decrypt(data)
        with open(f'decrypted_aes_cbc_{image_path}', 'wb') as file:
            file.write(decrypted_data)
```
